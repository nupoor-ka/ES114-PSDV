```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import math
```

```python
# DISCRETE UNIFORM DISTRIBUTION

begin = int(input('Enter lower limit of domain: '))
end = int(input('Enter upper limit of domain: '))
si = int(input('Enter number of data points to be generated: '))
vals = np.random.randint(begin, end, (si))
prob = 1/si
probs = np.full((si), prob)

fig = plt.figure(facecolor = 'white')
ax = fig.add_subplot()
ax.plot(vals, probs, color = 'green', linewidth = 0, marker = '.')
ax.bar(vals, probs, width = 1, color = 'lightblue')
plt.title('DISCRETE UNIFORM DISTRIBUTION', fontweight = 'bold')
plt.xlabel('VALUES OF X')
plt.ylabel('PROBABILITY')
plt.show()
```

```python
# BERNOULLI DISTRIBUTION

p = float(input('Enter probability of value 1: '))
q = 1 - p
x = np.array([1, 0])
y = np.array([p, q])

fig = plt.figure(facecolor = 'white')
ax = fig.add_subplot()
ax.plot(x, y, color = 'darkgreen', linewidth = 0, marker = '.')
ax.bar(x, y, width = 0.01, color = 'lightgreen')
plt.title('BERNOULLI DISTRIBUTION')
plt.xlabel('VALUES OF X')
plt.ylabel('PROBABILITY')
plt.show()
```

```python
# BINOMIAL DISTRIBUTION

p = float(input('Enter probability of success: '))
q = 1 - p
a = []
n = int(input('Enter number of times event takes place: '))
x = np.arange(1, n+1)
for i in range(1,n+1):
  a.append((math.comb(n, i))*(p**i)*(q**(n-i)))
y = np.array(a)

fig = plt.figure(facecolor = 'white')
ax = fig.add_subplot()
ax.plot(x, y, color = 'darkblue', linewidth = 0.2, marker = '.')
ax.bar(x, y, width = 0.1, color = 'lightblue')
plt.title('BINOMIAL DISTRIBUTION', fontweight = 'bold')
plt.xlabel('VALUES OF X')
plt.ylabel('PROBABILITY')
plt.show()
```

```python
# POISSON DISTRIBUTION

lm = int(input('Enter average rate of expected value: '))
t = int(input('Enter total time: '))
x = np.arange(1, lm*2)
a = []
for i in range(1, lm*2):
  p = lm**i
  q = np.exp(lm)
  r = math.factorial(i)
  a.append(p/(q*r))
y = np.array(a)

fig = plt.figure(facecolor = 'white')
ax = fig.add_subplot()
ax.plot(x, y, color = 'blue', linewidth = 0.4, marker = '.')
ax.bar(x, y, width = 0.5, color = 'lightblue')
plt.title('POISSON DISTRIBUTION', fontweight = 'bold')
plt.xlabel('VALUES OF X')
plt.ylabel('PROBABILITY')
plt.show()
```

```python
# GEOMETRIC DISTRIBUTION

p = float(input('Enter probability of success: '))
q = 1-p
n = int(1/p)+10
a = []
x = np.arange(1, n+1)
for i in range(1, n+1):
  a.append((q**(i-1))*p)
y = np.array(a)

fig = plt.figure(facecolor = 'white')
ax = fig.add_subplot()
ax.plot(x, y, color = 'lightblue', linewidth = 0.2, marker = '.')
ax.bar(x, y, width = 0.3, color = 'lightblue')
plt.title('GEOMETRIC DISTRIBUTION', fontweight = 'bold')
plt.xlabel('VALUES OF X')
plt.ylabel('PROBABILITY')
plt.show()
```

```python
# POISSON AND BINOMIAL DISTRIBUTION

n = 100
p = 0.15
lm = n*p
a1 = []
a2 = []

x1 = np.arange(0, int(lm*2))
for i in x1:
  p1 = lm**i
  q1 = np.exp(lm)
  r1 = math.factorial(i)
  a1.append(p1/(q1*r1))
y1 = np.array(a1)


x2 = np.arange(1, n+1)
q2 = 1 - p
for j in range(1,n+1):
  a2.append((math.comb(n, j))*(p**j)*(q2**(n-j)))
y2 = np.array(a2)

fig = plt.figure(facecolor = 'white')
ax = fig.add_subplot()
plt.xlabel('VALUES OF X')
plt.ylabel('PROBABILITY')

ax.plot(x1, y1, color = 'red', linewidth = 0.5, marker = '.')
ax.bar(x1, y1, width = 0.5, color = 'red')

ax.plot(x2, y2, color = 'lightblue', linewidth = 0.2, marker = '.')
ax.bar(x2, y2, width = 0.5, color = 'lightblue')

plt.title('POISSON AND BINOMIAL', fontweight = 'bold')
plt.show()
```
