```python
# LAB 4 - 23110224

import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
import math
```

1 (a)

```python
# UNIFORM DISTRIBUTION PDF

a = int(input('Enter beginning of range: '))
b = int(input('Enter end of range: '))
p = 1/(b-a)

if a == 0:
  x = []
  y = []
  yc = []
else:
  x = [0]
  y = [0]
  yc = [0]
x.extend([a, b, b+a])
y.extend([p, p, 0])


x = np.array(x)
y = np.array(y)
yc.extend([0, 1, 1])

plt.plot(x, y, marker = '', linewidth = 0)
plt.title('PDF - UNIFORM DISTRIBUTION')
plt.fill_between(x, y, where = y==p, color = 'lightblue')
plt.show()
```

```python
# UNIFORM DISTRIBUTION CDF

xc = np.array(x)
yc = np.array(yc)

plt.plot(xc, yc, marker = '', linewidth = 0)
plt.title('CDF - UNIFORM DISTRIBUTION')
plt.fill_between(xc, yc, where = xc!=0, color = 'lightblue')
plt.show()
```

```python
# GAUSSIAN / NORMAL DISTRIBUTION PDF

def pdf(x, mu, sig):
  return (1.0/math.sqrt(2*math.pi*sig**2)) * math.exp( -1 * (1.0/2.0) * ( (x - mu)/sig )**2 )
def func(number, colour):
  a = float(input('Enter the lower limit for plot '+number+': '))
  b = float(input('Enter the upper limit for plot '+number+': '))
  mu = float(input('Enter the mean for plot '+number+': '))
  sig = float(input('Enter the standard deviation for plot '+number+': '))
  x = list(np.linspace(a, b, 1000))
  y = []    # PDF
  yc = []   # CDF
  for i in x:
    y.append(pdf(i, mu, sig))
  yc = np.cumsum(y)
  plt.plot(x, y, marker = '', linewidth = 0.5, color = colour)
  plt.title('PDF - GAUSSIAN/NORMAL DISTRIBUTION')
  return x, yc

(x1, yc1) = func('1','red')
(x2, yc2) = func('2', 'blue')
```

2 (b)

```python
# GAUSSIAN / NORMAL DISTRIBUTION CDF

plt.plot(x1, yc1, marker = '', linewidth = 0.5, color = 'red')
plt.plot(x2, yc2, marker = '', linewidth = 0.5, color = 'blue')
plt.title('CDF - GAUSSIAN/NORMAL DISTRIBUTION')
```

```python
# EXPONENTIAL DISTRIBUTION PDF

def func(number, colour):
  lm = float(input('Enter the average number of events per unit time for plot '+number+': '))
  x = list(np.linspace(0, 5, 200))
  y = []    # PDF
  yc = []   # CDF
  for i in x:
    y.append(lm*(math.exp(-1*lm*i)))
    yc.append(1 - (math.exp(-1*lm*i)))
  plt.plot(x, y, marker = '', linewidth = 0.5, color = colour)
  plt.title('PDF - EXPONENTIAL DISTRIBUTION')
  return x, yc

(x1, yc1) = func('1','red')
(x2, yc2) = func('2', 'blue')
```

```python
# CAUCHY DISTRIBUTION PDF

def func(number, colour):

  x0 = float(input('Enter the median for plot '+number+': '))
  gam = float(input('Enter the scale parameter for plot '+number+': '))
  x = list(np.linspace(-5, 5, 200))
  y = []    # PDF
  yc = []   # CDF
  for i in x:
    b = (i - x0)/gam
    y.append(1/(3.14159*gam*(1 + b**2)))
    yc.append(0.5 + (math.atan(b))/3.14159)
  plt.plot(x, y, marker = '', linewidth = 0.5, color = colour)
  plt.title('PDF - CAUCHY DISTRIBUTION')
  return x, yc

(x1, yc1) = func('1','red')
(x2, yc2) = func('2', 'blue')
```

2 (d)

```python
# LAPLACHIAN DISTRIBUTION PDF

def func(number, colour):
  mu = float(input('Enter the location parameter for plot '+number+': '))
  b = float(input('Enter the scale parameter for plot '+number+': '))
  x = list(np.linspace(-10, 10, 300))
  y = []    # PDF
  yc = []   # CDF
  for i in x:
    c = (abs(i - mu)/b)
    y.append((math.exp(-1*c))/(2*b))
    if i<mu:
      yc.append(math.exp((i - mu)/b)/2)
    else:
      yc.append(1 - (math.exp(-1*((i - mu)/b))/2))
  plt.plot(x, y, marker = '', linewidth = 0.5, color = colour)
  plt.title('PDF - LAPLACIAN DISTRIBUTION')
  return x, yc

(x1, yc1) = func('1','red')
(x2, yc2) = func('2', 'blue')
```

2 (e)

```python
# LAPLACHIAN DISTRIBUTION CDF

plt.plot(x1, yc1, marker = '', linewidth = 0.5, color = 'red')
plt.plot(x2, yc2, marker = '', linewidth = 0.5, color = 'blue')
plt.title('CDF - LAPLACIAN DISTRIBUTION')
```

```python
# RAYLEIGH DISTRIBUTION PDF

def func(number, colour):
  sig = float(input('Enter the standard deviation for plot '+number+': '))
  x = list(np.linspace(0, 10, 300))
  y = []    # PDF
  yc = []   # CDF
  for i in x:
    c = math.exp(-0.5*((i/sig)**2))
    y.append((i*c)/(sig**2))
    yc.append(1 - c)
  plt.plot(x, y, marker = '', linewidth = 0.5, color = colour)
  plt.title('PDF - RAYLEIGH DISTRIBUTION')
  return x, yc

(x1, yc1) = func('1','red')
(x2, yc2) = func('2', 'blue')
```

2 (f)

```python
# RAYLEIGH DISTRIBUTION CDF

plt.plot(x1, yc1, marker = '', linewidth = 0.5, color = 'red')
plt.plot(x2, yc2, marker = '', linewidth = 0.5, color = 'blue')
plt.title('CDF - RAYLEIGH DISTRIBUTION')
```

```python
# DISCRETE UNIFORM DISTRIBUTION CDF



begin = int(input('Enter lower limit of domain: '))
end = int(input('Enter upper limit of domain: '))
si = int(input('Enter number of data points to be generated: '))
vals = np.random.randint(begin, end, (si))
vals = np.sort(vals)
x = list(vals)
prob = 1/si
cdf = []
for i in range(len(x)):
  cdf.append((i+1)*prob)

cdf = np.array(cdf)

fig = plt.figure(facecolor = 'white')
ax = fig.add_subplot()
ax.plot(vals, cdf, color = 'green', linewidth = 0, marker = '.')
ax.bar(vals, cdf, width = 1, color = 'lightblue')
plt.title('DISCRETE UNIFORM DISTRIBUTION CDF', fontweight = 'bold')
plt.xlabel('VALUES OF X')
plt.ylabel('CUMULATIVE DISTRIBUTION FUNCTION')
plt.show()
```

```python
# BERNOULLI DISTRIBUTION CDF

p = float(input('Enter probability of value 1: '))
q = 1 - p
x = np.array([0, 1])
y = np.array([q, p])
cdf = np.cumsum(y)

fig = plt.figure(facecolor = 'white')
ax = fig.add_subplot()
ax.plot(x, cdf, color = 'darkgreen', linewidth = 0, marker = '.')
ax.bar(x, cdf, width = 0.01, color = 'lightgreen')
plt.title('BERNOULLI DISTRIBUTION CDF')
plt.xlabel('VALUES OF X')
plt.ylabel('CUMULATIVE DISTRIBUTION FUNCTION')
plt.show()
```

```python
# BINOMIAL DISTRIBUTION CDF

p = float(input('Enter probability of success: '))
q = 1 - p
a = []
n = int(input('Enter number of times event takes place: '))
x = np.arange(1, n+1)
for i in range(1,n+1):
  a.append((math.comb(n, i))*(p**i)*(q**(n-i)))
y = np.array(a)
cdf = np.cumsum(y)


fig = plt.figure(facecolor = 'white')
ax = fig.add_subplot()
ax.plot(x, cdf, color = 'darkblue', linewidth = 0.2, marker = '.')
ax.bar(x, cdf, width = 0.1, color = 'lightblue')
plt.title('BINOMIAL DISTRIBUTION CDF', fontweight = 'bold')
plt.xlabel('VALUES OF X')
plt.ylabel('CUMULATIVE DISTRIBUTION FUNCTION')
plt.show()
```

```python
# POISSON DISTRIBUTION CDF

lm = int(input('Enter average rate of expected value: '))
t = int(input('Enter total time: '))
x = np.arange(1, lm*2)
a = []
for i in range(1, lm*2):
  p = lm**i
  q = np.exp(lm)
  r = math.factorial(i)
  a.append(p/(q*r))
y = np.array(a)
cdf = np.cumsum(y)

fig = plt.figure(facecolor = 'white')
ax = fig.add_subplot()
ax.plot(x, cdf, color = 'blue', linewidth = 0.4, marker = '.')
ax.bar(x, cdf, width = 0.5, color = 'lightblue')
plt.title('POISSON DISTRIBUTION CDF', fontweight = 'bold')
plt.xlabel('VALUES OF X')
plt.ylabel('CUMULATIVE DISTRIBUTION FUNCTION')
plt.show()
```

```python
# GEOMETRIC DISTRIBUTION CDF

p = float(input('Enter probability of success: '))
q = 1-p
n = int(1/p)+10
a = []
x = np.arange(1, n+1)
for i in range(1, n+1):
  a.append((q**(i-1))*p)
y = np.array(a)
cdf = np.cumsum(y)

fig = plt.figure(facecolor = 'white')
ax = fig.add_subplot()
ax.plot(x, cdf, color = 'lightblue', linewidth = 0.2, marker = '.')
ax.bar(x, cdf, width = 0.3, color = 'lightblue')
plt.title('GEOMETRIC DISTRIBUTION CDF', fontweight = 'bold')
plt.xlabel('VALUES OF X')
plt.ylabel('CUMULATIVE DISTRIBUTION FUNCTION')
plt.show()
```
